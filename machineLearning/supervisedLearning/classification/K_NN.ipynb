```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import neighbors
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
```

```python
sample = np.random.randint(70000, size=5000) # because mnist is too big for k-nn
data = mnist.data.iloc[sample]
target = mnist.target.iloc[sample]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(data, target, train_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xtest, ytest)
```

```python
error = 1 - knn.score(xtest, ytest)
print('Erreur: %f' % error)
```

```python
errors = []
for i in range(1, 10):
    knn = neighbors.KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain, ytrain)
    errors.append( 1 - knn.score(xtest, ytest))
plt.plot(range(1, 10), errors)
```
